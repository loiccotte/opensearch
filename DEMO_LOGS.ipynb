## Requêtes
### REQ 1 — Logs HTTP 404

````python
GET opensearch_dashboards_sample_data_logs/_search
{
  "query": {
    "term": { "response": 404 }
  }
}
````

**→ 801 résultats**
```json
{ "ip": "117.165.135.88", "response": 404, "request": "/opensearch_dashboards/opensearch_dashboards-1.0.0-linux-x86_64.tar.gz" }
```

### REQ 2 — Logs du mois de mai 2026

````python
GET opensearch_dashboards_sample_data_logs/_search
{
  "query": {
    "range": {
      "timestamp": { "gte": "2026-05-01", "lte": "2026-05-31" }
    }
  }
}
````

**→ 1866 résultats**
```json
{ "ip": "223.87.60.27", "response": 200, "timestamp": "2026-05-24T00:39:02.912Z" }
```

---
## Requêtes full-text
### REQ 3 — Logs taggés "error"

````python
GET opensearch_dashboards_sample_data_logs/_search
{
  "query": {
    "match": { "tags": "error" }
  }
}
````

**→ 829 résultats**
```json
{ "ip": "110.78.46.100", "response": 200, "tags": ["error", "info"] }
```

### REQ 4 — "opensearch" dans message ou URL

````python
GET opensearch_dashboards_sample_data_logs/_search
{
  "query": {
    "multi_match": {
      "query": "opensearch",
      "fields": ["message", "url"]
    }
  }
}
````

**→ 5101 résultats**
```json
{ "ip": "223.87.60.27", "url": "https://artifacts.opensearch.org/downloads/opensearch/opensearch-1.0.0.deb" }
```
