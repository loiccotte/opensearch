```python
from opensearchpy import OpenSearch

import pandas as pd
```

```python
client = OpenSearch(
    hosts=[{"host": "localhost", "port": 9200}],
    http_auth=("admin", "admin"),
    use_ssl=False,
    verify_certs=False,
    ssl_show_warn=False
)
```

```python
print(client.info())
```

```python
print(f'Statut : {client.cluster.health()["status"]}')



noeuds=client.cat.nodes(format="json")
print(noeuds)

```

```python
#Creation d'Index
config={
    "settings":{
        "index":{
            "number_of_shards":1, #Nb de fragments
            "number_of_replicas":2 #Copies de chaque shards
        }
    }
}

reponse=client.indices.create(index="mon_index",body=config,ignore=400)#ingore 400 evite erreur si index déja crée
print(reponse)
```
